```python
import pandas as pd
nba_fin = pd.read_excel('D:\\YouTube\\NBA\\NBA finals.xlsx')
nba_fin.head()
```

```python
nba_fin = nba_fin.drop(['Catg', 'Img'], axis = 1)
```

```python
players_to_plot = ['LeBron James', 'Stephen Curry', 'Michael Jordan', 'Jerry West']

start_year = '1947'
end_year = '2019'
```

```python
import matplotlib.pyplot as plt

first_year = '1947'
start_index = int(start_year) - int(first_year)
end_index = int(end_year) - int(first_year)

year = list(nba_fin)[1:]

x_value = year[start_index:end_index+1]
plt.figure(figsize=(7,5), dpi = 150)

for players in players_to_plot:
    players_df = nba_fin[nba_fin['Players'] == players]
    y_value = [int(players_df[col]) for col in x_value]
    plt.plot(x_value, y_value, label = players)

plt.legend()    
```
